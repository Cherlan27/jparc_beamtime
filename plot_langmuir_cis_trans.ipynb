```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#langmuir_trans = open('langmuir/test030601_05_fr121_d62DPPC_trans_processed.apd', mode="r")
#langmuir_cis = open('langmuir/test030601_05_fr121_d62DPPC_cis_processed.apd', mode="r")
langmuir_trans = open('langmuir/test030601_05_fr121_d62DPPC_NRW_trans_processed.apd', mode="r")
langmuir_cis = open('langmuir/test030601_05_fr121_d62DPPC_NRW_cis_processed.apd', mode="r")
amount = 25
```

```python
data_trans = langmuir_trans.readline()
data_cis = langmuir_cis.readline()
```

```python
pressure_cis = []
location_cis = []
pressure_trans = []
location_trans = []


data_cis_sep = data_cis.split(";")
data_trans_sep = data_trans.split(";")

for row in data_cis_sep:
    values = row.split(",")
    pressure_cis.append(float(values[5]))
    location_cis.append(float(values[8]))
location_cis = np.array(location_cis)
pressure_cis = np.array(pressure_cis)
for row in data_trans_sep:
    values = row.split(",")
    pressure_trans.append(float(values[5]))
    location_trans.append(float(values[8]))
location_trans = np.array(location_trans)
pressure_trans = np.array(pressure_trans)
```
